# 6. Data Visualization

Visual representations make it easier to identify trends, distributions, and patterns within the dataset.

The following charts summarize important characteristics of pesticide consumption in West Bengal.

## 6.1 Distribution of Pesticide Classes

This bar chart illustrates the frequency of different pesticide classes recorded in the dataset. It helps identify the dominant categories of pesticides used in agricultural practices.

```python
plt.figure(figsize=(8,5))

df["Pesticide_Class"].value_counts().plot(kind="bar")

plt.title("Distribution of Pesticide Classes")
plt.xlabel("Pesticide Class")
plt.ylabel("Frequency")
plt.xticks(rotation=45)

plt.show()
```

### Interpretation

The visualization highlights the relative occurrence of different pesticide classes within the dataset. The dominant class may indicate the primary agricultural challenges faced by farmers, such as weed control, insect management, or fungal disease prevention.

## 6.2 Distribution of WHO Toxicity Grades

The World Health Organization (WHO) classifies pesticides according to their toxicity. This chart summarizes the frequency of pesticides belonging to each toxicity category.

```python
plt.figure(figsize=(8,5))

df["WHO_Toxicity_Grade"].value_counts().plot(kind="bar")

plt.title("Distribution of WHO Toxicity Grades")
plt.xlabel("WHO Toxicity Grade")
plt.ylabel("Frequency")
plt.xticks(rotation=45)

plt.show()
```

### Interpretation

The dataset contains four WHO toxicity categories. The largest proportion of records belongs to **Not Classed (U)** with **51 observations (34.0%)**, followed closely by **Moderately Hazardous (Class II)** with **50 observations (33.3%)**. **Slightly Hazardous (Class III)** accounts for **35 observations (23.3%)**, while **Highly Hazardous (Class Ib)** is the least represented, with **14 observations (9.3%)**.

Although highly hazardous pesticides represent a relatively small fraction of the dataset, their presence highlights the continued use of chemicals that require strict handling, storage, and regulatory oversight. The predominance of unclassified and moderately hazardous pesticides suggests that agricultural practices in the dataset rely primarily on compounds considered less acutely toxic than Class Ib pesticides. Nevertheless, responsible pesticide management remains essential to minimize environmental contamination and protect human health.

## 6.3 Distribution of Crops

This chart presents the frequency of crop types represented in the dataset, indicating the agricultural systems included in the analysis.

```python
plt.figure(figsize=(10,5))

df["Crop_Cultivated"].value_counts().plot(kind="bar")

plt.title("Distribution of Crops")
plt.xlabel("Crop")
plt.ylabel("Frequency")
plt.xticks(rotation=45)

plt.show()
```

## 6.4 Most Frequently Recorded Pesticide Chemicals

This visualization displays the ten most frequently occurring pesticide chemicals in the dataset.

```python
plt.figure(figsize=(12,5))

df["Chemical_Name"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Pesticide Chemicals")

plt.xlabel("Chemical Name")

plt.ylabel("Frequency")

plt.xticks(rotation=60)

plt.show()
```

## 6.5 Pesticide Usage Across Agricultural Seasons

Agricultural seasons influence pest pressure and pesticide application patterns. This chart compares the number of observations across seasons.

```python
plt.figure(figsize=(8,5))

df["Agricultural_Season"].value_counts().plot(kind="bar")

plt.title("Agricultural Season Distribution")

plt.xlabel("Season")

plt.ylabel("Frequency")

plt.xticks(rotation=30)

plt.show()
```

## 6.6 Zone-wise Distribution

This chart compares the number of observations recorded across different geographical zones of West Bengal.

```python
plt.figure(figsize=(8,5))

df["Zone"].value_counts().plot(kind="bar")

plt.title("Zone-wise Distribution")

plt.xlabel("Zone")

plt.ylabel("Frequency")

plt.xticks(rotation=30)

plt.show()
```

## 6.7 Relationship Between Area Treated and Pesticide Usage

Scatter plots help visualize relationships between two numerical variables.

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Estimated_Area_Treated_K_Ha"],
    df["Total_Technical_Grade_Used_Tonnes"]
)

plt.title("Area Treated vs Total Technical Grade Used")

plt.xlabel("Estimated Area Treated (K Ha)")

plt.ylabel("Total Technical Grade Used (Tonnes)")

plt.show()
```

## 6.8 Correlation Analysis

Correlation analysis measures the strength and direction of relationships among numerical variables.

```python
corr = df.select_dtypes(include="number").corr()

plt.figure(figsize=(8,6))

plt.imshow(corr)

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.colorbar()

plt.title("Correlation Matrix")

plt.show()
```

```python
# Integrated Environmental Data Dashboard

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 4, figsize=(22, 10))

# -------------------------
# 1. Pesticide Class
# -------------------------
df["Pesticide_Class"].value_counts().plot(
    kind="bar",
    ax=axes[0,0],
    title="Pesticide Classes"
)
axes[0,0].tick_params(axis='x', rotation=45)

# -------------------------
# 2. WHO Toxicity
# -------------------------
df["WHO_Toxicity_Grade"].value_counts().plot(
    kind="bar",
    ax=axes[0,1],
    title="WHO Toxicity Grade"
)
axes[0,1].tick_params(axis='x', rotation=45)

# -------------------------
# 3. Crop Distribution
# -------------------------
df["Crop_Cultivated"].value_counts().plot(
    kind="bar",
    ax=axes[0,2],
    title="Crop Distribution"
)
axes[0,2].tick_params(axis='x', rotation=45)

# -------------------------
# 4. Agricultural Season
# -------------------------
df["Agricultural_Season"].value_counts().plot(
    kind="bar",
    ax=axes[0,3],
    title="Agricultural Season"
)
axes[0,3].tick_params(axis='x', rotation=30)

# -------------------------
# 5. Consumption Intensity
# -------------------------
axes[1,0].hist(
    df["Consumption_Intensity_KG_Per_Ha"],
    bins=12
)
axes[1,0].set_title("Consumption Intensity")

# -------------------------
# 6. Area Treated
# -------------------------
axes[1,1].hist(
    df["Estimated_Area_Treated_K_Ha"],
    bins=12
)
axes[1,1].set_title("Area Treated")

# -------------------------
# 7. Scatter Plot
# -------------------------
axes[1,2].scatter(
    df["Estimated_Area_Treated_K_Ha"],
    df["Total_Technical_Grade_Used_Tonnes"]
)
axes[1,2].set_title("Area vs Pesticide Used")
axes[1,2].set_xlabel("Area")
axes[1,2].set_ylabel("Tonnes")

# -------------------------
# 8. Correlation Matrix
# -------------------------
corr = df.select_dtypes(include="number").corr()

im = axes[1,3].imshow(corr)

axes[1,3].set_xticks(range(len(corr.columns)))
axes[1,3].set_xticklabels(corr.columns, rotation=90)

axes[1,3].set_yticks(range(len(corr.columns)))
axes[1,3].set_yticklabels(corr.columns)

axes[1,3].set_title("Correlation Matrix")

fig.colorbar(im, ax=axes[1,3])

plt.suptitle(
    "Integrated Environmental Data Analysis Dashboard",
    fontsize=18,
    fontweight="bold"
)

plt.tight_layout()

plt.show()
```

# 7. AI Prompt Engineering

## Objective

Prompt Engineering is the practice of designing clear, specific, and context-rich instructions that enable Artificial Intelligence (AI) systems to generate accurate, relevant, and high-quality outputs.

Unlike traditional search engines, Generative AI models rely heavily on the quality of the prompt. A well-structured prompt provides the AI with a defined role, sufficient context, explicit analytical tasks, and the desired response format.

In this project, Prompt Engineering is applied to transform a general-purpose AI model into an Environmental Data Analyst capable of interpreting pesticide consumption data and generating scientifically meaningful insights.

## Components of the Prompt

The prompt was designed using the following principles:

- Role Assignment
- Dataset Context
- Analytical Objectives
- Expected Output Structure
- Scientific Writing Style
- Environmental Perspective

# 8. Prompt Used


You are an Environmental Data Scientist specializing in agricultural sustainability, environmental toxicology, and data analytics.

You have been provided with a pesticide consumption dataset from West Bengal.

Your objective is to analyze the dataset and prepare a scientific environmental assessment.

Your report should include:

1. Executive Summary
2. Dataset Overview
3. Key Statistical Findings
4. Major Trends and Patterns
5. Environmental Implications
6. Agricultural Implications
7. Human Health Considerations
8. Data Quality Assessment
9. Recommendations for Sustainable Pesticide Management
10. Suitable Machine Learning Models
11. Future Research Opportunities
12. Final Conclusion

Use formal scientific language.

Explain every important observation using evidence from the dataset.

Avoid generic statements.

Where appropriate, relate pesticide usage to environmental sustainability.

# 9. AI-Generated Environmental Assessment

## Objective

The structured prompt developed in the previous section was provided to ChatGPT (GPT-5.5) to generate an AI-assisted environmental assessment of the pesticide consumption dataset.

The following assessment is based on the uploaded dataset, exploratory data analysis, and the integrated environmental dashboard generated during this project.

## 9.1 Executive Summary

The pesticide consumption dataset comprises **150 observations** collected between **2018 and 2026**, representing agricultural pesticide usage across multiple crops, pesticide classes, toxicity grades, agricultural seasons, and geographical zones in West Bengal.

Exploratory Data Analysis (EDA) revealed considerable variability in pesticide application intensity, treated agricultural area, and total technical-grade pesticide usage. The average treated agricultural area was **66.74 thousand hectares**, while the mean pesticide consumption intensity was **0.414 kg/ha** and the average technical-grade pesticide consumption was **27.52 tonnes**.

The integrated dashboard indicates that **insecticides** constitute the largest pesticide class represented in the dataset. The WHO toxicity profile is dominated by **Not Classed (U)** and **Moderately Hazardous (Class II)** pesticides, whereas **Highly Hazardous (Class Ib)** pesticides account for a comparatively smaller proportion of observations. The scatter plot further demonstrates a positive relationship between treated agricultural area and total pesticide consumption, while the correlation matrix confirms meaningful associations among selected numerical variables.

Overall, the dataset reflects diverse pesticide management practices across different agricultural systems. The combination of statistical analysis, visualization, and AI-assisted interpretation provides a comprehensive understanding of pesticide usage patterns and highlights opportunities for more sustainable agricultural management.

## 9.2 Dataset Overview

The dataset was analyzed using Python within the Google Colab environment. Data manipulation and statistical analysis were performed using the Pandas and NumPy libraries, while Matplotlib was used to generate visualizations.

The dataset contains both categorical and numerical variables describing pesticide usage characteristics, including:

- Year of observation
- Crop cultivated
- Agricultural season
- Geographical zone
- Pesticide class

- WHO toxicity grade

The dataset spans nine years (2018–2026), enabling the exploration of pesticide application patterns across multiple agricultural contexts. No significant structural issues were identified that would prevent exploratory analysis, and the dataset proved suitable for AI-assisted interpretation.

## 9.3 Statistical Interpretation

Descriptive statistical analysis indicates substantial variation in pesticide application practices throughout the study period.

The estimated agricultural area treated ranged from **5.03 thousand hectares** to **119.35 thousand hectares**, with an average value of **66.74 thousand hectares**, suggesting considerable differences in cultivation scale among observations.

Pesticide consumption intensity varied between **0.113 kg/ha** and **0.999 kg/ha**, with a mean value of **0.414 kg/ha**. This variability indicates that pesticide application rates differ across crops, seasons, and management practices.

Similarly, total technical-grade pesticide usage ranged from **1.64 tonnes** to **103.40 tonnes**, averaging **27.52 tonnes**. The relatively high standard deviations observed for these variables indicate substantial heterogeneity within the dataset, reflecting the diverse agricultural conditions represented.

Overall, the descriptive statistics demonstrate that the dataset captures a broad spectrum of pesticide usage scenarios, providing a robust foundation for exploratory analysis and AI-assisted environmental assessment.

## 9.4 Key Findings

The AI-assisted assessment identified several important patterns within the pesticide consumption dataset.

The integrated dashboard indicates that insecticides constitute the largest pesticide class represented in the dataset, suggesting that insect pest management is a major component of agricultural practices in West Bengal. Fungicides and herbicides are also widely represented, whereas biopesticides account for a comparatively smaller proportion of observations.

Analysis of the WHO toxicity categories shows that the dataset is dominated by **Not Classed (U)** and **Moderately Hazardous (Class II)** pesticides, accounting for the majority of observations. Slightly Hazardous (Class III) pesticides are moderately represented, while Highly Hazardous (Class Ib) pesticides constitute the smallest category. Although the proportion of highly hazardous pesticides is relatively low, their continued use warrants careful regulatory oversight and safe handling practices.

The distribution of observations across Kharif, Rabi, and Zaid seasons demonstrates that pesticide application occurs throughout the agricultural calendar rather than being concentrated within a single cropping season. Similarly, multiple crop types are represented, indicating diverse agricultural systems with varying pest management requirements.

The scatter plot demonstrates a positive relationship between the estimated agricultural area treated and the total technical-grade pesticide used. This suggests that larger cultivation areas generally require greater pesticide inputs, although variability among observations indicates that factors such as crop type, pest pressure, and management practices also influence pesticide consumption.

## 9.5 Environmental Implications

The results highlight several environmental considerations associated with pesticide use.

The widespread occurrence of synthetic pesticides indicates their continued importance in maintaining agricultural productivity. However, repeated application of these compounds may contribute to environmental challenges including soil contamination, surface and groundwater pollution, impacts on non-target organisms, and reductions in biodiversity if not managed responsibly.

The presence of moderately and highly hazardous pesticides further emphasizes the importance of adopting integrated pest management (IPM), precision agriculture, and environmentally sustainable pesticide application strategies. Increasing the use of biological control methods and lower-risk alternatives may help reduce ecological impacts while maintaining crop productivity.

Overall, the findings demonstrate the need to balance agricultural production with long-term environmental sustainability through evidence-based pesticide management.

## 9.6 Agricultural Implications

From an agricultural perspective, the dataset reflects considerable diversity in pesticide application across crops, seasons, and cultivation systems.

The variability observed in pesticide consumption intensity suggests that application practices are influenced by multiple factors including crop characteristics, pest infestation levels, climatic conditions, and farm management strategies.

These findings reinforce the importance of crop-specific pesticide recommendations, precision application technologies, and continuous farmer education to optimize pesticide use while minimizing unnecessary chemical inputs. Data-driven decision-making can improve productivity while supporting sustainable agricultural development.

## 9.7 Human Health Considerations

Although the dataset is primarily focused on agricultural pesticide usage, the toxicity profile provides valuable insight into potential human health considerations.

The relatively low proportion of Highly Hazardous (Class Ib) pesticides is encouraging from a public health perspective. Nevertheless, the continued presence of hazardous pesticide categories highlights the importance of safe handling procedures, appropriate personal protective equipment (PPE), proper storage, and adherence to recommended application guidelines.

Strengthening awareness among farmers and agricultural workers can reduce occupational exposure while improving food safety and environmental protection.

## 9.8 Recommendations for Sustainable Pesticide Management

Based on the AI-assisted assessment, the following recommendations are proposed:

- Promote Integrated Pest Management (IPM) to reduce dependence on chemical pesticides.
- Encourage the adoption of biological pesticides and environmentally friendly alternatives where feasible.
- Improve farmer training on safe pesticide handling, storage, and application practices.

- Utilize digital agriculture technologies and data-driven decision support systems to optimize pesticide application.
- Strengthen regulatory monitoring of highly hazardous pesticides to minimize environmental and human health risks.
